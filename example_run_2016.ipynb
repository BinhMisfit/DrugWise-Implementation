```python
import pandas as pd
from sklearn.ensemble import GradientBoostingRegressor
from scipy.stats import pearsonr
```

```python
def train_test(df): 
    core_key = pd.read_csv("./utils/pdb2016/cor_key_2016.csv")
    cor_names = core_key['pro_name']
    gen_key = pd.read_csv("./utils/pdb2016/gen_key_2016.csv")
    gen_names = gen_key['pro_name']
    test_df = df[df['pro_name'].isin(cor_names)]
    train_df = df[df['pro_name'].isin(gen_names)]
    return test_df, train_df
```

```python
# replace with data in onedrive link
df = pd.read_csv(f'./2016/L_5_6_0.5_general_2016_sum_pw_pkd.csv', index_col = 0) 
df_test, df_train = train_test(df)
```
